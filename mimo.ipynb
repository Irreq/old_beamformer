```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

mimo_data = np.load("data/mimo.npy")

# customizing plot
plt.title("MIMO Beamforming")

pixel_plot = plt.imshow(
  mimo_data, interpolation='none', extent=[-60, 60,-60, 60])
plt.colorbar(pixel_plot)
  
# save a plot
plt.savefig('out/pixel_plot.png')
  
# show plot
plt.show(pixel_plot)
```
